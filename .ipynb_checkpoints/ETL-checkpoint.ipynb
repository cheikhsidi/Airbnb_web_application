```python
#fin#fill NaN data with 0 in the dataframe and display the data
df.fillna('0',inplace=True)
df
```

```python
df["host_is_superhost"]  = df["host_is_superhost"].str.replace('t', 'Yes').str.replace('f', 'No')
```
